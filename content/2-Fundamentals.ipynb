## Objects: a minimal introduction

### Class, attributes and methods (what is the _point_?)

Python is a language based in _objects_. This means that is easier to understand if we have a minimal knowledge of what an object is.

An object is a structure, like any other object in nature. It is composed form parts (called _attributes_) and functions (called _methods_). Imagine your friend's car. You can write a table of all the different attributes methods that makes the car a useful object:

__Attributes:__ pneumatics (x4), steering wheel, lights, gear ...

__Methods:__ to break, to accelerate, to move forward, to move backwards ...


 _Objects_ are also called _instances_ and the process of defining an object is _instantiation_. The creation (instantiation) of your friend's car as a particular instance of the class car will read in python as

```python
my_friends_car=car(parameters)
```

where _parameters_ is a series of data defining the different _attributes_ (the brand of the pneumatics, the type of lights, etc.). Some of these parameters define the particular _attributes_ of your friend's car. Once you have _my_friends_car_ instantiated, you can access to them using a _point_:

```python
print(my_friends_car.pneumatics)
```

and you can drive the car using methods, that you will call also using the _point_

```python
my_friends_car.accelerate()
```

#### Lopping over lists

Python allows to loop directly over a container, whithout using an auxiliary index. A `for` loop looks as this
  
```python
ex1=[0.2,3,4.5,'+','+',0.3,-4,8, '-', '-', 1]
for n in ex1:  # n is not an index, loops directly the values stored in the list
    print(n)
    
# 0.2
# 3
# 4.5
# .
# .
# .
```
Note in comparison with other programmmng languages that:
- There is no need to use an index as loop variable
- The loop starts with a `for`sentence ended by `:``
- The loop content is always indented (displaced to the right).
- There is no _end loop_ sentence, as the loop content is recognized by the indentation.

Sometimes, however, it is desirable to know the position of every item  during the loop. We access to this information adding an `enumerate()` statement

```python
ex1=[0.2,3,4.5,'+','+',0.3,-4,8, '-', '-', 1]
for i, n in enumerate(ex1):
    print(i, n)

# 0 0.2
# 1 3
# 2 4.5
# .
# .
# .
```

### Positional and key and default arguments.

Consider a simple function

```python
def my_function(a,b):
    print(a+2*b)

my_function(2,1)

# 4
```
Paramenters `a`and `b` are positional argumets, since in calling he function the first argument is always interpreted as `a`and the second as`b`. If we invert the position during the function call then the values foor `a`and `b` will be interchanged

```python
def my_function(a,b):
    print(a+2*b)

my_function(1,2) # calling the function with two positional arguments

# 5
```

The function also requires both arguments, if one is not given it will crash (try it!).

In addition to positional arguments, in the function call `a` and `b` can be defined by its keys. In such case, they can be located in any order:

```python
def my_function(a,b):
    print(a+2*b)
    
my_function(a=2,b=1)     #  calling the function with two key arguments
my_function(b=1, a=2)    # calling the function with two key arguments with changed poositions

# 4
# 4
```

> **NOTE:** Calling a function with key arguments is better, much clearer and avoids mistakes!

In the function definition, we can give default values to the arguments, so in case one of the arguments is not defined, the function will take the default.

```python
def my_function(a=2,b=3):
    print(a+2*b)
    
my_function(b=1)     #  calling the function with missing a, will assume the default a=2
my_function(a=5)     #  calling the function with missing b, will assume the default b=3
my_function()        #  calling the function with the two missing arguments, will assumne the defaults a=2 and b=3

# 4
# 11
# 8
```

### *args and **kwargs

Sometimes the number or arguments in a function can be changed depending on the particular situation when calling it. For instance, we may want a function that adds all the argumets, therefore with the following calls and results:

sum(2,1) $\longrightarrow$ 3

sum(2,1,3) $\longrightarrow$ 6

sum(2,1,3,0,9,1) $\longrightarrow$ 16

For this, in the function implementation, the optional function parameters will be stored in a _tuple_ called `args`. A _tuple_ is like a list but it cannot be changed.

```python
def sum(a, *args):
    if len(args) !=0:
        print('inside the function the optional arguments are the list args=',args)
        sum=a
        for iarg in args:
            sum=sum+iarg
        return sum
    else:
        return a

result=sum(2,1); print(result)
result=sum(2,1,3); print(result)
result=sum(2,1,3,0,9,1); print(result)

# inside the function the optional arguments are the list args= (1,)
# 3
# inside the function the optional arguments are the list args= (1, 3)
# 6
# inside the function the optional arguments are the list args= (1, 3, 0, 9, 1)
# 16
```

The same situation can happen for a variable list of arguments given with a key, instead of being positional. In this case the key arguments are parsed into the function as a dictionary called `kwargs`

```python
def sum(first, **kwargs):
    if len(kwargs) !=0:
        print('inside the function the optional arguments are a dictionary kwargs=',kwargs)
        sum=first
        for iarg in kwargs.values():
            sum=sum+iarg
        return sum
    else:
        return first

result=sum(2,a=1); print(result)
result=sum(2,b=1,a=3); print(result)
result=sum(2,b=1,a=3,d=0,c=9,f=1); print(result)

# inside the function the optional arguments are a dictionary kwargs= {'a': 1}
# 3
# inside the function the optional arguments are a dictionary kwargs= {'b': 1, 'a': 3}
# 6
# inside the function the optional arguments are a dictionary kwargs= {'b': 1, 'a': 3, 'd': 0, 'c': 9, 'f': 1}
# 16
```

## Classes and objects.

We have seen that "almost" everyting in Pyhton is an object. Now we shall learn how to define our own objects.

To define an object, you first need to define its structure, which we call a __class__.

For example, consider an object corresponding to a vector in a two-dimensional space. When defining a class, it is essential to define an initialization method, named __init__.

The most basic definition of our vector class would be as follows:

```python
class vector:
    
    def __init__(self,x,y): # initialization method. 
        self.x=x # self represents the object's name and 
                 # self.x is the attribute x (x coordinate)
        self.y=y # defining y coordinate as another attribute
```

Now that we have defined the class, we can generate (instantiate) various objects of the class "vector"
```python
v1=vector(1,2)
v2=vector(-1,1)

print('Vector v1 has coordinates x= ', v1.x, ' and y= ', v1.y)
print('Vector v2 has coordinates x= ', v2.x, ' and y= ', v2.y)
```

Let us now extend our class 'vector' to include some _methods_ (i.e. functions) like the scalar product:

```python
class vector:
    
    def __init__(self,x,y): # initialization method. 
        self.x=x # self represents the object's name and 
                 # self.x is the attribute x (x coordinate)
        self.y=y # defining y coordinate as another attribute
    
    def prod_with(self,u): # dot product with vector u (also an object of the vector class)
        prod=self.x*u.x+self.y*u.y
        return prod

v1=vector(1,2)
v2=vector(-1,1)

print('The scalar product of v1 times v2 is ', v1.prod_with(v2))
print('The modulus squared of v1 is the scalar product with itself = ', v1.prod_with(v1))
```

### _Dunders_

One thing that stands out in our previous example of class definition is that one of the methods (_init_) is surrounded by two underscores \_\_ on each side. These types of methods are called _dunders_ (double underscore). They function exactly as any other method, the reason for enclosing them with underscores is a convention: they are internal methods of the class and they are  not intended to be used directly by the user.

One interesting use of dunders is for defining arithmetic operations in a class. For example, in our vector class, we have defined the dot product with the method prod_with(). It would be very useful if we could perform the dot product using the * operator. This is called _operator overloading_ and it is easy to program by simply changing the method's name to the dunder __mul__.

```python
class vector:
    
    def __init__(self,x,y): # initialization method. 
        self.x=x # self represents the object's name and 
                 # self.x is the attribute x (x coordinate)
        self.y=y # defining y coordinate as another attribute
    
    def __mul__(self,u): # dot product with vector u (also an object of the vector class)
        prod=self.x*u.x+self.y*u.y
        return prod

v1=vector(1,2)
v2=vector(-1,1)

print('The scalar product of v1 times v2 is ', v1*v2)
print('The modulus squared of v1 is the scalar product with itself = ', v1*v1)
```

### Modules

Just to finish this lesson and in order that you have a minimum functionality to create python scripts, let us have a word on python _modules_.

_Modules_ are the python name for _libraries_: a collection of functions, variables and object definitions that you can load from disk an that increase the functionality of the basic python language.

To import a module we have to specify at the begining of our script an `import` sentence. For instance, very important to us and the subject of the next lesson is the mathematical module called _numpy_.

To import _numpy_ just add the following sentence at the top of your script:

```python
import numpy as np
```
this tells the interpreter to  load the _numpy_ library and to call it with the alias _np_. 

This, as any other module, is treated as as an object. It has methods (the mathematical functions), and attribvutes (the mathematical constants).

See this script as a simple example

```python
import numpy as np

a=np.pi # a equals the attribute (mathematical constant) pi

print(np.sin(a)) # calls the trigonometric sin method and applies it to a

# 1.2246467991473532e-16
``` 

Many times it is unjustified to load a whole library if one wants to use just some methods. To avoid this, we can specify the interpreter which metghos are we going to use nd how to call it:

```python
from  numpy import sin as my_sin, pi as my_pi

a=my_pi # a equals the attribute (mathematical constant) pi

print(my_sin(a)) # calls the trigonometric sin method and applies it to a

# 1.2246467991473532e-16
```

Probably you are as surprised as I was: changing the list `a` changes _magically_ list `b`! This is dangerous if you believe that `b` is an unaltered copy of the initial `a`, a situation very frequent when programming.

Before trying to _correct_ this behaviour and force python not to change `b` when changing `a` we must now what is going on (looks like a bug...).

> __NOTE:__ The following information may be a bit advanced. If you want to skip it is OK. You can jump to the last subsection: [Wrap up](#wrapup), and continue from there.

#### Variables referring to a simple objects (numbers, etc.)

In many languages __variables are locations in memory__, where you can store numbers, etc. 
Memory is a sort of _chest of drawers_ and each variable is a _drawer_ where you can store copies of them.

Variable assignation in python follows a different perspective. __Numbers, etc are stored only in one place in memory. Variables contain a label to them.__ 


Instead of a _chest of drawers_, memory in python is a _shelf_ with objects, and variables only store the refernece to them, not the object itself.

> __NOTE:__ If you are familiar to _pointers_, variables in python are _pointers_ that point to the same place in memory.

In the following example, `a`and `b` refer to the same object in the memory shelf. If we write `a=3`, we change the reference to the object `3` in the shelf.

<img src="2-Fundamentals_images/variables_python.png" alt="Description of image"
     width="900" height="600"
     style="object-fit: contain; display: block; margin: auto; background-color: white;"/>

So far... so good. Although the way of dealing with variables is different, this does not affect our programming, as it gives the same result in python and in FORTRAN.

```python
a=2
b=a
print('a= ',a, 'b= ',b)

a=3
print('a= ',a, 'b= ',b)

 # a=  2 b=  2
 # a=  3 b=  2
```

You can always check if two variables refer to the same object looking for the _label_ of their content. For instance.

```python
a=2
b=1
print(id(a), id(b) # they refer to different objects 
      
b=a
print(id(a), id(b)) # now a and b point to the same object
print(id(2)) # the object labelled is number 2!

# 4561103184 4561103152
# 4561103184 4561103184
# 4561103184
```

#### Variables referring to complex objects (lists, dictionaries, objects, etc.)


Things, however, are not going to be the same when storing complex structures, like containers (list, dictionaries,...) or objects.

Take a list, and let us do the same experiment above:


```python
a=[1,2,3]; print('a= ',a ) # a is a list
b=a; print('b= ',b ) # b is a list as a

a[0]=4; print('a= ',a ) # change the value of the first element in a

#a=  [1, 2, 3]
#b=  [1, 2, 3]
#a=  [4, 2, 3]
#b=  [4, 2, 3]
```

As you can see, it looks like changing `a` changes also `b`. But remember that when we assign two variables as `a=b` in pyhton both variables refer to the same object. We can check this by looking for the object identifier

```python
a=[1,2,3]
b=a

print(id(a), id(b))

# 4609709632 4609709632
```

#### Use `copy` to replicate complex objects!

So, if we want to _duplicate_ a list, instead of using `=` we have to use the list method `.copy()`

```python
a=[1,2,3]; print('a= ',a ) # a is a list
b=a.copy(); print('b= ',b ) # b is a list as a

a[0]=4; print('a= ',a ) # change the value of the first element in a

print('b= ',b)  # b is not changed anymore
print(id(a), id(b)) # a and b point to different objevt in memory

# a=  [1, 2, 3]
# b=  [1, 2, 3]
# a=  [4, 2, 3]
# b=  [1, 2, 3]
# 4609715072 4609407936
``` 


#### Use `deepcopy` to replicate complex objects composed from other complex objects!

It is possible that your list can have complex objevct in it. For instande a lists that have other lists as elements. For instance:

```python
list1=[[1,2],3,4]
```
 We say the the  `list1` is at first level, and `[1,2]`at a second (deepper) level. The method `.copy()` in `list1` will duplicate the first level, but not the second. To see this clearer, look to the following example
 
 ```python
list1=[[1,2],3,4]; print(list1)
list2=list1.copy()
list1[0][1]=10

print(list1, list2)  # list1[0][1] and list2[0][1] are changed
print(id(list1), id(list2)) # although at the first level list1 and list2 are duplicated (not the same object)
print(id(list1[0]), id(list2[0])) # the list at the deeper level is still the same 
 # [[1, 2], 3, 4]
 # [[1, 10], 3, 4] [[1, 10], 3, 4]
 # 4609717248 4609716672
 # 4609714624 4609714624
```


We can duplicate a variable at all levels using the `deepcopy()` function. This is not a function of basic python, so it should be imported from the module `copy``

```python
from copy import deepcopy

list1=[[1,2],3,4]; print(list1)
list2=deepcopy(list1)
list1[0][1]=10

print(list1, list2)  # list2[0][1] is not changed
print(id(list1), id(list2)) # at the first level list1 and list2 are duplicated (not the same object)
print(id(list1[0]), id(list2[0])) # at the the deeper level they are also duplicated! 

# [[1, 2], 3, 4]
# [[1, 10], 3, 4] [[1, 2], 3, 4]
# 4609718080 4609701824
# 4609347328 4609715456
```



####  Wrap up

In python assignment 'a=b' does not duplicate the value of `a` and stores it in `b`, as many other languages do. `a=b` meand that `a`and `b`point to the same object in memoory: they are identical.

This charcateristic of python is not important when `a` and `b` are variables refering to simple objects (numbers, strings, etc), _but is makes a difference_ when `a` and `b` refer to structures as lists.

_RECEIPT:_

- To duplicate simple structures, like a list of numbers, never use `=`. Instead use the method `.copy()`, as in this example:
```python
a=[1,2,3]
b=a.copy()
print(a,b)
```

- To duplicate complex structures, like lists of lists, yoou use the function `deepcopy()` that must be imported from the module `copy`. Look to this example:

```python
from copy import deepcopy
a=[[11,22],2,3]
b=deepcopy(a)
print(a,b)
```

> __NOTE:__ Though you can use `deepcopy()` instead of the method `.copy()` to copy simple structures, `deepcopy()`is much slower!
